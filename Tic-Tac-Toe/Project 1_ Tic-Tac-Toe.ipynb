```python
from IPython.display import clear_output
import random
```

```python
# Create display board

def display(test_board):
  clear_output()
  print(test_board[7] + "|" + test_board[8] + "|" + test_board[9])
  print(test_board[4] + "|" + test_board[5] + "|" + test_board[6])
  print(test_board[1] + "|" + test_board[2] + "|" + test_board[3])
```

```python
# Take input from user if they want to be 'O' or 'X'

def player_input():
  marker = ""
  while not(marker == "X" or marker == "O"):
    marker = input("Player 1: Do you want to be X or O").upper()

  if marker == 'X':
    return ('X','O')
  else:
    return ('O','X')
```

```python
# Fill the user value on board

def place_marker(board, marker, position):
  board[position] = marker
```

```python
# Winning goals

def win_check(board, mark):
  return ((board[7]==mark and board[8]==mark and board[9]==mark) or #across the top
          (board[4]==mark and board[5]==mark and board[6]==mark) or #across the middle
          (board[1]==mark and board[2]==mark and board[3]==mark) or #across the bottom
          (board[7]==mark and board[4]==mark and board[1]==mark) or #across the left side
          (board[2]==mark and board[5]==mark and board[8]==mark) or #across the middle
          (board[3]==mark and board[6]==mark and board[9]==mark) or #across the right side
          (board[7]==mark and board[5]==mark and board[3]==mark) or #across the diagonal
          (board[9]==mark and board[5]==mark and board[1]==mark))
```

```python
# If space is empty

def space_check(board, position):
  return board[position]==" " 
```

```python
# Decide whose turn is first

def choose_first():
  if random.randint(0,1) == 0:
    return 'Player 2'
  else:
    return 'Player 1'
```

```python
# if board is already full

def full_board_check(board):
  for i in range(1, 10):
    if space_check(board, i):
      return False
  return True
```

```python
# Take the input from user

def player_choice(board):
  position = 0
  while position not in range(1,10) or not space_check(board, position):
    position = int(input("\n\nEnter your position(0-9) "))

  return position
```

```python
# If user want to play it again

def replay():
  return input("You want to play again(yes, no)").lower().startswith('y')
```

```python
# Compile all

print("Welcome to Tic-Tac-Teo!!!")

while True:
  theBoard = [' ']*10
  player1_marker, player2_marker = player_input()
  turn = choose_first()
  print(turn + " will go first") 

  play_game = input("Do you want to play it again(Yes/No)")

  if play_game.lower()[0] =='y':
    game_on = True
  else:
    game_on = False

  while game_on:
    if turn == 'Player 1':

      display(theBoard)
      position = player_choice(theBoard)
      place_marker(theBoard, player1_marker, position)

      if win_check(theBoard, player1_marker):
        display(theBoard)
        print("Congratulations! You have won the game")
        game_on = False

      else:
        if full_board_check(theBoard):
          display(theBoard)
          print("The game is draw!")
          break
        else:
          turn = "Player 2"

    else:

      display(theBoard)
      position = player_choice(theBoard)
      place_marker(theBoard, player2_marker, position)

      if win_check(theBoard, player2_marker):
        display(theBoard)
        print("Congratulations! You have won the game")
        game_on = False

      else:
        if full_board_check(theBoard):
          display(theBoard)
          print("The game is draw!")
          break
        else:
          turn = "Player 1"
          
  if not replay():
    break
```
